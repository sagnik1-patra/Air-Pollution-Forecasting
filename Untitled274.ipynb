```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
import pickle
import yaml
import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Conv1D, MaxPooling1D, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# ================================================================
#  PATHS
# ================================================================
BASE = r"C:\Users\NXTWAVE\Downloads\Air Pollution Forecasting"
FILES = [
    os.path.join(BASE, "archive", "station_hour.csv"),
    os.path.join(BASE, "archive", "stations.csv"),
    os.path.join(BASE, "archive", "city_hour.csv"),
    os.path.join(BASE, "archive", "station_day.csv"),
    os.path.join(BASE, "archive", "city_day.csv"),
]

# ================================================================
#  LOAD AND MERGE DATA
# ================================================================
print("[INFO] Loading datasets...")
df_hour_station = pd.read_csv(FILES[0])
df_stations = pd.read_csv(FILES[1])
df_hour_city = pd.read_csv(FILES[2])
df_day_station = pd.read_csv(FILES[3])
df_day_city = pd.read_csv(FILES[4])

df = df_day_station.copy()

# ================================================================
#  DATA PREPROCESSING
# ================================================================
print("[INFO] Preprocessing data...")
df = df.fillna(method='ffill').fillna(method='bfill')
drop_cols = ['station_id', 'station_name', 'Unnamed: 0']
for c in drop_cols:
    if c in df.columns:
        df.drop(c, axis=1, inplace=True, errors='ignore')

df = df.select_dtypes(include=[np.number])
target_col = 'PM2.5' if 'PM2.5' in df.columns else df.columns[-1]
X = df.drop(columns=[target_col])
y = df[target_col]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.values.reshape(-1, 1))
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=42
)

# ================================================================
#  AIS (Stage 1  Exploration)
# ================================================================
print("[INFO] Stage 1: AIS Exploration...")

def evaluate_model(params):
    filters, lstm_units, dropout_rate, lr = params
    model = Sequential([
        Conv1D(int(filters), kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
        MaxPooling1D(2),
        LSTM(int(lstm_units), return_sequences=False),
        Dense(32, activation='relu'),
        Dropout(dropout_rate),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    hist = model.fit(X_train, y_train, epochs=8, batch_size=32, validation_split=0.2, verbose=0)
    return np.min(hist.history['val_loss'])

# AIS exploration  generate immune population
population_size = 6
population = [
    [random.choice([32, 64, 128]), random.choice([32, 64, 128]), random.uniform(0.1, 0.4), random.uniform(0.0005, 0.005)]
    for _ in range(population_size)
]
fitness = [evaluate_model(p) for p in population]

for gen in range(5):
    print(f"[AIS] Generation {gen+1} | Best Fitness: {min(fitness):.6f}")
    for i in range(population_size):
        if random.random() < 0.6:
            population[i][0] = random.choice([32, 64, 128])
            population[i][1] = random.choice([32, 64, 128])
            population[i][2] = max(0.1, min(0.5, population[i][2] + random.uniform(-0.05, 0.05)))
            population[i][3] = max(0.0001, min(0.01, population[i][3] + random.uniform(-0.0003, 0.0003)))
    fitness = [evaluate_model(p) for p in population]

best_params_ais = population[np.argmin(fitness)]
print("[AIS] Best parameters:", best_params_ais)

# ================================================================
#  CSA (Stage 2  Refinement)
# ================================================================
print("[INFO] Stage 2: CSA Refinement...")

def levy_flight(Lambda):
    u = np.random.normal(0, 1)
    v = np.random.normal(0, 1)
    step = u / (abs(v) ** (1 / Lambda))
    return step

best_params = best_params_ais.copy()
best_score = evaluate_model(best_params)

for i in range(5):
    new_params = best_params.copy()
    new_params[0] = max(32, min(128, new_params[0] + int(levy_flight(1.5) * 10)))
    new_params[1] = max(32, min(128, new_params[1] + int(levy_flight(1.5) * 10)))
    new_params[2] = max(0.1, min(0.5, new_params[2] + levy_flight(1.5) * 0.05))
    new_params[3] = max(0.0001, min(0.01, new_params[3] + levy_flight(1.5) * 0.001))
    
    score = evaluate_model(new_params)
    if score < best_score:
        best_score = score
        best_params = new_params
    print(f"[CSA] Iteration {i+1} | Best Fitness: {best_score:.6f}")

print("[CSA] Final optimized parameters:", best_params)

# ================================================================
#  Final CNN + LSTM Model
# ================================================================
filters, lstm_units, dropout_rate, lr = best_params
model = Sequential([
    Conv1D(int(filters), kernel_size=3, activation='relu', input_shape=(X_train.shape[1], 1)),
    MaxPooling1D(2),
    LSTM(int(lstm_units), return_sequences=False),
    Dense(32, activation='relu'),
    Dropout(dropout_rate),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(
    X_train, y_train,
    epochs=50, batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# ================================================================
#  Evaluation
# ================================================================
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
results = {"MAE": float(mae), "RMSE": float(rmse), "R2": float(r2)}
print(json.dumps(results, indent=4))

# ================================================================
#  Save Outputs (Sequential Prefix)
# ================================================================
os.makedirs(BASE, exist_ok=True)
model.save(os.path.join(BASE, "sequential_air_forecasting_model.h5"))
with open(os.path.join(BASE, "sequential_air_forecasting_results.json"), "w") as f:
    json.dump(results, f, indent=4)
with open(os.path.join(BASE, "sequential_air_forecasting_config.yaml"), "w") as f:
    yaml.dump(
        {
            "optimizer": "Sequential Hybrid (AIS  CSA)",
            "best_params": {
                "filters": int(filters),
                "lstm_units": int(lstm_units),
                "dropout_rate": float(dropout_rate),
                "learning_rate": float(lr),
            },
        },
        f,
    )
with open(os.path.join(BASE, "sequential_air_forecasting_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

# ================================================================
#  VISUALIZATIONS (Display + Save)
# ================================================================

# 1 Accuracy / Loss Graph
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs'); plt.ylabel('MSE')
plt.legend(); plt.grid(True)
plt.tight_layout(); plt.show()
plt.savefig(os.path.join(BASE, "sequential_air_forecasting_loss_graph.png"))
plt.close()

# 2 Prediction Graph
plt.figure(figsize=(7,6))
plt.scatter(y_test, y_pred, color='royalblue', alpha=0.6)
plt.xlabel('Actual PM2.5'); plt.ylabel('Predicted PM2.5')
plt.title('Predicted vs Actual PM2.5 Concentration')
plt.grid(True); plt.tight_layout(); plt.show()
plt.savefig(os.path.join(BASE, "sequential_air_forecasting_prediction_graph.png"))
plt.close()

# 3 Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.tight_layout(); plt.show()
plt.savefig(os.path.join(BASE, "sequential_air_forecasting_heatmap.png"))
plt.close()

# 4 Comparison Graph
plt.figure(figsize=(6,4))
labels = list(results.keys()); values = list(results.values())
plt.bar(labels, values, color=['skyblue','lightgreen','salmon'])
plt.title('Sequential Hybrid AIS+CSA  Model Metrics')
plt.ylabel('Score'); plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout(); plt.show()
plt.savefig(os.path.join(BASE, "sequential_air_forecasting_comparison_graph.png"))
plt.close()

# 5 Result Graph
plt.figure(figsize=(10,5))
plt.plot(y_test[:200], label='Actual', color='blue')
plt.plot(y_pred[:200], label='Predicted', color='orange', linestyle='--')
plt.title('Sequential Hybrid  Result Graph (First 200 Samples)')
plt.xlabel('Samples'); plt.ylabel('Scaled PM2.5')
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
plt.savefig(os.path.join(BASE, "sequential_air_forecasting_result_graph.png"))
plt.close()

print("[] All sequential hybrid graphs displayed and saved successfully at:")
print(BASE)
```
